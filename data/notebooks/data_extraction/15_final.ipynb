```python
"""
PIPELINE STAGE: Final Multi-Dimensional Master Dataset Synthesis
INPUTS: 
    - 13_energy_population_geo_day.xlsx (Panel Data with Energy, Demographics, and Solar metrics)
    - 14_altitude.xlsx (Topographic Reference Dataset)
OUTPUT: 01_final.xlsx
LIBRARIES: pandas, os

1. RESEARCH CORE: RELATIONAL DATA BROADCASTING & FEATURE FUSION
    The ultimate objective of this script is to synthesize a high-resolution 
    multi-dimensional master dataset. By integrating static topographic markers 
    (Altitude) into a dynamic longitudinal framework, the script creates a 
    comprehensive 'Feature Space' ready for machine learning and econometric modeling.

2. TECHNICAL SPECIFICATIONS & ROBUSTNESS:
    - Case-Insensitive Schema Mapping: Implements a proactive normalization layer 
      to resolve naming inconsistencies across different datasets (e.g., handling 
      'Year', 'Yıl', 'yil' as 'year'). This ensures zero-loss during relational joins.
    - Relational Key Enforcement: Ensures that the 'plate' (Province ID) is 
      strictly cast as an integer across all sources to prevent common data 
      mismatches during the 'Left Join' operation.
    - Dimensional Scaling: Correctly broadcasts time-invariant topographic 
      features (Altitude) across the temporal axis (2020-2024), enriching every 
      spatio-temporal observation.

3. DATA GOVERNANCE & SORTING:
    - Chronological Alignment: Multi-level hierarchical sorting by [Plate], [Year], 
      and [Month] is enforced to preserve the time-series structure essential 
      for panel data analysis.
    - Error Handling & Diagnostics: Real-time exception catching with a dynamic 
      column-mapping readout to troubleshoot any schema drift in source files.

4. ACADEMIC SIGNIFICANCE:
    This final synthesis eliminates the silos between energy consumption, 
    demographic density, geospatial coordinates, solar duration, and topographic 
    elevation. The result is a unified, analytical-grade dataset optimized for 
    predicting regional energy demand variances.
"""

import pandas as pd
import os

def final_master_birlesimi_v2():
    # 1. Dosya Yolları
    MASTER_PATH = os.path.join("..", "..", "processed_data", "steps", "13_energy_population_geo_day.xlsx")
    ALTITUDE_PATH = os.path.join("..", "..", "processed_data", "steps", "14_altitude.xlsx")
    FINAL_OUT = os.path.join("..", "..", "processed_data", "steps","15_final.xlsx")

    try:
        # 2. Dosyaları Yükle
        print("Dosyalar yükleniyor...")
        df_master = pd.read_excel(MASTER_PATH)
        df_alt = pd.read_excel(ALTITUDE_PATH)

    
        # 4. Veri Tiplerini Sabitle
        df_master['Plate'] = df_master['Plate'].astype(int)
        df_alt['Plate'] = df_alt['Plate'].astype(int)

        # 5. Birleştirme (Left Join)
        print("Rakım verileri (Altitude) ana tabloya ekleniyor...")
        final_df = pd.merge(
            df_master, 
            df_alt[['Plate', 'Altitude']], 
            on='Plate', 
            how='left'
        )

        # 6. Sıralama (Sort)
        # Artık 'year' ve 'month' sütunlarının küçük harf olduğundan eminiz
        print("Veri seti plaka ve zaman ekseninde sıralanıyor...")
        final_df = final_df.sort_values(by=['Plate', 'Year', 'Month']).reset_index(drop=True)

        # 7. Kaydet
        final_df.to_excel(FINAL_OUT, index=False)
        
        print("\n" + "="*60)
        print("TEBRİKLER: MAKİNE ÖĞRENMESİ MASTER VERİ SETİ TAMAMLANDI")
        print(f"Dosya Adresi: {FINAL_OUT}")
        print(f"Toplam Gözlem (Satır): {len(final_df)}")
        print(f"Sütun Sayısı: {len(final_df.columns)}")
        print("="*60)

    except Exception as e:
        print(f"\n!!! HATA OLUŞTU: {e}")
        print("Master tablonun mevcut sütunları şunlar:")
        if 'df_master' in locals():
            print(df_master.columns.tolist())

if __name__ == "__main__":
    final_master_birlesimi_v2()
```
