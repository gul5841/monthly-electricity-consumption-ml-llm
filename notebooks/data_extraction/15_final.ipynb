```python
"""
PIPELINE STAGE: Final Multi-Dimensional Master Dataset Synthesis
PERIOD: January 2020 - December 2024
LIBRARIES: os, pandas

1. RESEARCH OBJECTIVE: RELATIONAL BROADCASTING & FEATURE FUSION
   The ultimate objective of this script is to synthesize a high-resolution, 
   multi-dimensional master dataset. By integrating static topographic markers (Altitude) 
   into a dynamic longitudinal framework, the script finalizes a comprehensive 
   'Feature Space' ready for advanced econometric modeling and machine learning algorithms.

2. FILE ARCHITECTURE:
   - Input 1 (Master Panel): os.path.join("..", "..", "processed_data", "steps", "13_energy_population_geo_day.xlsx")
   - Input 2 (Topography): os.path.join("..", "..", "processed_data", "steps", "14_altitude.xlsx")
   - Output (ML Ready): os.path.join("..", "..", "processed_data", "steps", "15_final.xlsx")

3. TECHNICAL SPECIFICATIONS & ROBUSTNESS:
   - Relational Key Enforcement: Ensures that the 'Plate' (Province ID) is strictly 
     cast as an integer across all sources to prevent data mismatches and guarantee 
     absolute precision during the 'Left Join' operation.
   - Selective Dimensional Broadcasting: Pulls ONLY the 'Altitude' column from the 
     topographic dataset and correctly broadcasts this time-invariant feature across 
     the entire temporal axis (2020-2024), enriching every spatio-temporal observation 
     without causing schema bloat.

4. DATA GOVERNANCE & SORTING:
   - Chronological Alignment: Multi-level hierarchical sorting by ['Plate', 'Year', 'Month'] 
     is rigorously enforced to preserve the strict time-series structure essential 
     for longitudinal panel data analysis.
   - Diagnostics & Error Handling: Implements a dynamic exception-catching block that 
     automatically dumps the Master DataFrame's column list to the console upon failure, 
     allowing researchers to rapidly troubleshoot schema drift.

5. ACADEMIC SIGNIFICANCE:
   This final synthesis definitively eliminates the silos between energy consumption, 
   demographic density, geospatial coordinates, solar photoperiods, and topographic 
   elevation. The result is a unified, analytical-grade matrix optimized for predicting 
   regional energy demand variances.
"""
import pandas as pd
import os

def final_master_birlesimi_v2():
    # 1. Dosya Yolları
    MASTER_PATH = os.path.join("..", "..", "processed_data", "steps", "13_energy_population_geo_day.xlsx")
    ALTITUDE_PATH = os.path.join("..", "..", "processed_data", "steps", "14_altitude.xlsx")
    FINAL_OUT = os.path.join("..", "..", "processed_data", "steps","15_final.xlsx")

    try:
        # 2. Dosyaları Yükle
        print("Dosyalar yükleniyor...")
        df_master = pd.read_excel(MASTER_PATH)
        df_alt = pd.read_excel(ALTITUDE_PATH)

    
        # 4. Veri Tiplerini Sabitle
        df_master['Plate'] = df_master['Plate'].astype(int)
        df_alt['Plate'] = df_alt['Plate'].astype(int)

        # 5. Birleştirme (Left Join)
        print("Rakım verileri (Altitude) ana tabloya ekleniyor...")
        final_df = pd.merge(
            df_master, 
            df_alt[['Plate', 'Altitude']], 
            on='Plate', 
            how='left'
        )

        # 6. Sıralama (Sort)
        # Artık 'year' ve 'month' sütunlarının küçük harf olduğundan eminiz
        print("Veri seti plaka ve zaman ekseninde sıralanıyor...")
        final_df = final_df.sort_values(by=['Plate', 'Year', 'Month']).reset_index(drop=True)

        # 7. Kaydet
        final_df.to_excel(FINAL_OUT, index=False)
        
        print("\n" + "="*60)
        print("TEBRİKLER: MAKİNE ÖĞRENMESİ MASTER VERİ SETİ TAMAMLANDI")
        print(f"Dosya Adresi: {FINAL_OUT}")
        print(f"Toplam Gözlem (Satır): {len(final_df)}")
        print(f"Sütun Sayısı: {len(final_df.columns)}")
        print("="*60)

    except Exception as e:
        print(f"\n!!! HATA OLUŞTU: {e}")
        print("Master tablonun mevcut sütunları şunlar:")
        if 'df_master' in locals():
            print(df_master.columns.tolist())

if __name__ == "__main__":
    final_master_birlesimi_v2()
```
